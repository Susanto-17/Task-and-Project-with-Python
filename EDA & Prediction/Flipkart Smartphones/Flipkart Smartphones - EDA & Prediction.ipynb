```python
x = xiaomi_smartphone.drop(['brand', 'model', 'base_color', 'processor', 'screen_size', 'display_size', 'num_rear_camera', 'num_front_camera', 'battery_capacity', 'sales'], axis = 1)

y = xiaomi_smartphone['sales']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
print('Training Accuracy :', model.score(x_train, y_train))
print('Testing Acccuracy :', model.score(x_test, y_test))
```
